```python
from bs4 import BeautifulSoup
import requests
```

```python
path = "https://en.wikipedia.org/wiki/List_of_best-selling_mobile_phones"
```

```python
requests.get(path)
page = requests.get(path)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
info_table = soup.find_all('table')[1]
```

```python
info_table
```

```python
info_table.find_all('th')
```

```python
headers = info_table.find_all('th')
```

```python
print(headers)
```

```python
headers = [head.text.strip() for head in headers]
```

```python
print(headers)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=headers)
```

```python
df
```

```python
raw_data = info_table.find_all('tr')
```

```python
for data in raw_data:
    print(len(data))
```

```python
for row in raw_data:
    tds = row.find_all('td')

    if len(tds) == 7:
        selected_tds = tds[:5] + [tds[5]]
    else:
        selected_tds = tds
    individual_row = [tds.text.strip() for tds in selected_tds]
    if len(individual_row) == len(headers):
        df.loc[len(df)] = individual_row
```

```python
manufacturer_table = soup.find_all('table')[4]
```

```python
column_names = manufacturer_table.find_all('th')
```

```python
column_names = [name.text.strip() for name in column_names]
```

```python
print(column_names)
```

```python
manu_df = pd.DataFrame(columns=column_names)
```

```python
manu_df.head()
```

```python
manu_data = manufacturer_table.find_all('tr')
```

```python
manu_data
```

```python
for data in manu_data:
    row = data.find_all('td')
    individual_row = [td.text.strip() for td in row]
    if len(individual_row) == len(column_names):
        cleaned_row = [val if val else None for val in individual_row]
        manu_df.loc[len(manu_df)] = cleaned_row
```
